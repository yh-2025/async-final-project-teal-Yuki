```python
''' Question 3 Who buys the most wine based on education levels?

The "Education" column of the dataset has the following entries: "Basic', "Graduation", "2n Cycle", "Master", and "PhD"
    * due to personal research, concluded that basic means did not graduate high school, graduation means high school graduate and 2n cycle means undergrad college graduate
'''

# Group and get top 5 education levels by wine consumption
most_sold = marketing_df.groupby('Education')['MntWines'].sum().nlargest(5)

# Prepare data for pie chart
labels = most_sold.index        
sizes = most_sold.values

changing_labels = list(labels)
changing_labels[0] = changing_labels[0] + " (High School Graduate)"
changing_labels[4] = changing_labels[4] + " (No High School Diploma)"
changing_labels[3] = changing_labels[3] + " (Undergrad College Graduate)"


# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=changing_labels, autopct='%1.1f%%', startangle=140)
plt.title("Top Wine Consumers by Education Level")
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.show()

```
