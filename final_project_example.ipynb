```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset. Replace 'titanic.csv' with your own dataset file for your project.
# Ensure the dataset file exists and is in the same directory as this jupyter notebook.
marketing_df = pd.read_csv('marketing.csv',sep='\t')

# Display the first few rows of the dataframe to ensure it is loaded correctly
display(marketing_df.head())

#inspect dataset
marketing_df.info()
marketing_df.describe()
display(marketing_df[marketing_df["Income"].isna()])

```

```python
# Question 1 Does the adverage
#  amount of complaints change if you have kids or no kids?


## Create a new column, where if Kidhome is more than 0, it's True meaning haskids. If not, then put False
marketing_df["haskids"] = (marketing_df["Kidhome"] > 0 )

#Calculate the complaints

## FIRST TRY: complain or not based on having kids
complaint_avg = marketing_df.groupby('haskids')['Complain'].mean()

## SECOND TRY: Complain: did customer complain in last 2 years. 1:True, 0:False
## haskids: True if they have at least 1 kid, false is 0 kids
haskids_complain_df = marketing_df.groupby("Complain")['haskids'].sum()

display(haskids_complain_df)

print(marketing_df["Complain"].unique())
sns.barplot(data=marketing_df, x="haskids", y="Complain")
plt.title("Number of complaints change if you have kids ")
plt.xticks([0,1], ["No Kids", "Has Kids"])
plt.ylabel("Average Complaints")
plt.show()
```

```python
#Question 2 Which campaign out of the 5 did the most people purchase from?



campaign_comp = marketing_df.groupby('')[""].max()
sns.barplot(campaign_comp)
plt.title("Top campaign")
plt.xlabel("campaigns")
plt.xticks(rotation=45)
plt.ylabel("campaign sum")
plt.show()
```

```python
''' Question 3 Who buys the most wine based on education levels?

The "Education" column of the dataset has the following entries: "Basic', "Graduation", "2n Cycle", "Master", and "PhD"
    * due to personal research, concluded that basic means did not graduate high school, graduation means high school graduate and 2n cycle means undergrad college graduate
'''

# Group and get top 5 education levels by wine consumption
most_sold = marketing_df.groupby('Education')['MntWines'].sum().nlargest(5)

# Prepare data for pie chart
labels = most_sold.index        
sizes = most_sold.values

changing_labels = list(labels)
changing_labels[0] = changing_labels[0] + " (High School Graduate)"
changing_labels[4] = changing_labels[4] + " (No High School Diploma)"
changing_labels[3] = changing_labels[3] + " (Undergrad College Graduate)"


# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=changing_labels, autopct='%1.1f%%', startangle=160)
plt.title("Top Wine Consumers by Education Level")
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.show()

```

```python
#Question 5 What is the adverage amount of gold bought by people in relationships compared to people that are not?

avg_online = marketing_df["NumStorePurchases"].mean()
avg_instore = marketing_df["NumWebPurchases"].mean()

gold_bought = marketing_df.groupby('Marital_Status')['MntGoldProds'].mean()

### drop "YOLO" and "Alone" and "Absurd"
filtered_gold_bought = gold_bought.drop(["YOLO", "Alone", "Absurd"])

plt.figure(figsize=(8, 6))
plt.pie(filtered_gold_bought.values, labels=filtered_gold_bought.index, autopct='%1.1f%%', startangle=160)
plt.title("Relationship status compared to gold purchases")
plt.axis('equal')  # Ensures pie is drawn as a circle
plt.show()
```

```python
#Question 6 What do consumers buy more of vegetables or sweets?

avg_online = marketing_df["NumStorePurchases"]
avg_instore = marketing_df["NumWebPurchases"]


plo_df = pd.DataFrame({
    "Product": ["Online", "In Store"],
    "Average Spending": [avg_online, avg_instore]
})

# Plot it
sns.barplot(data=plo_df, x="Product", y="Average Spending")
plt.title("Average Spending: Online vs In Store")
plt.xlabel("Product Type")
plt.ylabel("Average Amount Spent")
plt.show()
```

```python
#Question 7 Do consumers like to shop on adverage online or in stores?

avg_online = marketing_df["NumStorePurchases"].mean()
avg_instore = marketing_df["NumWebPurchases"].mean()


plo_df = pd.DataFrame({
    "Place of Purchase": ["Online", "In Store"],
    "Average Spending": [avg_online, avg_instore]
})

# Plot it
sns.barplot(data=plo_df, x="Place of Purchase", y="Average Spending")
plt.title("Average Spending: Online vs In Store")
plt.xlabel("place of purchase")
plt.ylabel("Average Amount Spent")
plt.show()
```

```python
#Question 8 What is the adverage of deals used compared to income levels?

marketing_df["cleaned_income"] = (marketing_df["Income"].dropna())
most_deals = marketing_df.groupby('NumDealsPurchases')['Income'].mean()

sns.scatterplot(most_deals)
plt.title("Adverage deals used by Income levels")
plt.xlabel("Number of deals used")
plt.ylabel("Income")
plt.show()
```
